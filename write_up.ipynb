## Introduction

Our goal was to find potential high-value players, that we should recruit to our team, and determine key identifying features of potential high-value candidates. We were able to do this using salary and statistical data collected from the 2025-2026 NBA season. Key variables that we looked at for this project were: total # of field-goals made (FG), Salary, total # of rebounds (TRB), total # of assists (AST), total # of steals (STL), total # of blocks (BLK), and # of games played (G). It is important to note that we kept player names as a label, but it was not used to train the model. All data was scaled in relation to the minimum and maximum of each respective statistic. What that means, is each statistical value is represented as a percentage of the maximum value of that statistical category. Our K-means algorithm was optimized to categorize the NBA players into four groups. The mean of each statistical category within each cluster are represented in the table below. 

![Average Stats Tabel](avg_stats.png)

Cluster 3 is going to be your highest valued players, having all statistical production above 30% of the potential statistical maximum, their salary averages only 17% of the statistical maximum, and they play an average of 89% of games per season. These players have an overall balanced performance and low-salary demand making them the highest valued players. Players in cluster 1 are considered your high-ticket all-star talent. These players are expensive, with some performing up to those standards. However, in our situation, I would suggest against going after talent like this, as it takes up cap space, and hinders our ability to go after higher valued talent in cluster 3. Our next best under-valued group is going to be cluster 0. These players might not have the highest statistical production, however, they can come off the bench consistently, giving your starters time to rest, and stay healthy. Moving on, I highly advise you to stay away from players in Cluster 2. Players in this group have comparable performance to the all-stars in group 1, with an average salary that is 10% of the statistical maximum, but they have a noticeably low average games played. These players cannot be relied on for consistent performance.

## Top Picks

Through graphical analysis, I have come up with 3 lists of players, to give you examples of top prospects, next best options, and players to stay away from. Starting with the top talent, Amen Thompson, Kon Knueppel, and Nickeil Alexander-Walker are 3 examples of players that should be at the top of our list. Their salaries are all below 0.26, meaning they are only being paid, at most, 26% of the statistical salary maximum. Conversely, they all above 45% the statistical maximum in at least 2 productivity categories and were able to keep this productivity up for at least 90% of the season. Consistent and affordable players, like these, are key to building a championship calaber team. A great rule of thumb, if a player has a salary at, or below 25% of the statistical maximum, and has at least two stats above 45% of the statistical maximum, all while playing 90% or more games in a season, they are should be at the top of your list.
